# Insights

## Represening the different users

```python
columns = df.columns

# Initialize lists to store weekly metrics
new_users = []
retained_users = []
resurrected_users = []
churned_users = []
quick_ratios = []
total_users = []

# Convert the data into a list of sets for easier comparison
weekly_active_users = [set(df[col]) for col in columns]

# Calculate metrics for each week
for i in range(len(weekly_active_users)):
    total_users.append(len(weekly_active_users[i]))
    if i == 0:
        new_users.append(len(weekly_active_users[i]))
        retained_users.append(0)
        resurrected_users.append(0)
        churned_users.append(0)
    else:
        previous_weeks_union = set().union(*weekly_active_users[:i])
        new_users.append(len(weekly_active_users[i] - previous_weeks_union))
        retained_users.append(len(weekly_active_users[i] & weekly_active_users[i - 1]))
        resurrected_users.append(len(weekly_active_users[i] & previous_weeks_union - weekly_active_users[i - 1]))
        churned_users.append(len(weekly_active_users[i - 1] - weekly_active_users[i]))

# Calculate quick ratios
for i in range(len(weekly_active_users)):
    if churned_users[i] == 0:
        quick_ratios.append(float('inf'))  # Avoid division by zero
    else:
        quick_ratios.append((new_users[i] + resurrected_users[i]) / churned_users[i])

# Create a DataFrame to store the results
results_df = pd.DataFrame({
    'Week': columns,
    'New Users': new_users,
    'Retained Users': retained_users,
    'Resurrected Users': resurrected_users,
    'Churned Users': churned_users,
    'Quick Ratio': quick_ratios,
    'Total Users': total_users
})
results_df
```

## Observation

**The total number of users are the same. The quick ratio is constant for all the weeks**

## New Users vs Weeks

```python
plt.figure(figsize=(14, 7))

# Plot New Users
plt.plot(results_df['Week'], results_df['New Users'], label='New Users', marker='o', color='blue')

# Adding labels and title
plt.xlabel('Weeks')
plt.ylabel('Number of Users')
plt.title('Weekly New Users')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

# Show the line graph
plt.tight_layout()
plt.show()
```

## Observation

**There is a rapid decline in the number of new users joining the app. The number of users is maximum at the start and then it rapidly decreases**

## Retained Users vs Weeks

```python
plt.figure(figsize=(14, 7))
plt.plot(results_df['Week'], results_df['Retained Users'], label='Retained Users', marker='o', color='green')
plt.xlabel('Weeks')
plt.ylabel('Number of Users')
plt.title('Weekly Retained Users')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observation

**The number of retained users is consistantly high. The majority of customer base for the company is fixed set of returning users. This shows that company has good product or service. So the decline in new users is not due to decrease in the quality of product or service. The shows that people are unaware of the company.There is need for marketing to reach the company to the users**

## Resurrected Users vs Weeks

```python
plt.figure(figsize=(14, 7))
plt.plot(results_df['Week'], results_df['Resurrected Users'], label='Resurrected Users', marker='o', color='orange')
plt.xlabel('Weeks')
plt.ylabel('Number of Users')
plt.title('Weekly Resurrected Users')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observation

**We can clearly see that the resurrected users spike once every 4 to 5 weeks and then dips. This can be necause the service provided by the company is such that monitoring it once a month is sufficient**

## Churned Users vs Weeks

```python
plt.figure(figsize=(14, 7))
plt.plot(results_df['Week'], results_df['Churned Users'], label='Churned Users', marker='o', color='red')
plt.xlabel('Weeks')
plt.ylabel('Number of Users')
plt.title('Weekly Churned Users')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observation

**We can clearly see that when churned users spike, it immediate decreases. This shows that whenevever there is an issue with the product or service and users leave, there is prompt response from the company to solve the issue and re-invite the users. This indicates very good customer support and problem solving from the company. This maybe the main reason why the company has such a strong retained user base**

## Quick Ratio vs Weeks

```python
plt.figure(figsize=(14, 7))
plt.plot(results_df['Week'], results_df['Quick Ratio'], label='Quick Ratio', marker='o', color='purple')

# Adding labels and title
plt.xlabel('Weeks')
plt.ylabel('Quick Ratio')
plt.title('Weekly Quick Ratio')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## Observation

**A constant quick ratio proves that the company is progressing very stably. This can be attributed to the retained users base due to good service by the company. The company is doing well

## Summary
